```python
from database import initialize_database
import sqlite3

def add_user_types():
    try:
        #connecting to the database
        with initialize_database() as conn:
            cursor = conn.cursor()

            #inserting user types into UserType table
            user_types = '''
                INSERT INTO UserType
                (typeName)
                VALUES
                ('Guest'),
                ('Member'),
                ('Waiter'),
                ('Manager'),
                ('Owner');
            '''
            cursor.execute(user_types)
            conn.commit()
            print("User types added successfully")
        #handling exceptions
    except sqlite3.OperationalError:
        print('Nie można otworzyć pliku bazy danych. Sprawdź uprawnienia dostępu.\nZamykanie programu.')

add_user_types()
```
